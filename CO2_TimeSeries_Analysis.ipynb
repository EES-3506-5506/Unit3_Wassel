The purpose of this notebook was to create a plot of CO2 concentrations from Mauna Loa, NOAA, and WDCGG on

```python
import pandas as pd                          
import matplotlib.pyplot as plt 
import numpy as np
```

```python
%matplotlib inline                           
%config InlineBackend.figure_format='retina' 
plt.rcParams['figure.figsize'] = 12,6  
```

```python
CO2_MaunaLoa = pd.read_csv('./data/greenhouse_gases/CO2_MaunaLoa_TimeSeries.csv',parse_dates=['Date'])
CO2_NOAA = pd.read_csv('./data/greenhouse_gases/CO2_NOAA_TimeSeries.csv',parse_dates=['Date'])
CO2_WDCGG = pd.read_csv('./data/greenhouse_gases/CO2_WDCGG_TimeSeries.csv',parse_dates=['Date'])
```

```python
CO2_MaunaLoa
```

```python
CO2_MaunaLoa.plot(x='Date',
            y='Mauna Loa CO2 (ppm)',
            linewidth=2)
```

```python
CO2_NOAA
```

```python
CO2_NOAA.plot(x= 'Date',
              y= 'NOAA CO2 (ppm)',
              linewidth=2)
```

```python
CO2_WDCGG
```

```python
CO2_WDCGG.plot(x= 'Date',
               y= 'WDCGG CO2 (ppm)',
               linewidth= 2)
```

```python
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(CO2_MaunaLoa['Date'],
         CO2_MaunaLoa['Mauna Loa CO2 (ppm)'],
         linewidth=2,label='Mauna Loa',
         color='red')

## Set properties of title, ylabels and xlabels
ax.set_title('Measured CO2 from Mauna Loa, NOAA, WDCGG',
             fontsize=12, fontweight='bold')
ax.set_ylabel('CO2 concentration [ppm]',
             fontsize=12, fontweight='bold')
ax.set_xlabel('Year',
             fontsize=12, fontweight='bold')
ax.plot(CO2_NOAA['Date'],
        CO2_NOAA['NOAA CO2 (ppm)'],
        linewidth= 2, label= 'NOAA',
        color= 'yellow')
ax.plot(CO2_WDCGG['Date'],
        CO2_WDCGG['WDCGG CO2 (ppm)'],
        linewidth= 2, label= 'WDCGG',
        color='blue')
plt.legend()
```

Graph of CO2 Concentrations on the y axis from Mauna Loa, NOAA, and WDCGG plotted against the year on the x axis. Mauna Loa data is in red, NOAA's data is in yellow, and WDCGG's data is in blue.
